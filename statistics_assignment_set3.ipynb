## Q1: Difference between t-test and z-test

**Answer:**
- z-test: Used when population SD is known and sample size is large
- t-test: Used when population SD is unknown and sample size is small

Example: Use t-test for small classroom data

## Q2: One-tailed vs Two-tailed test

**Answer:**
- One-tailed: Tests in one direction
- Two-tailed: Tests in both directions

## Q3: Type 1 and Type 2 errors
